```python
for ind, row in frankfurt_set.iterrows():
    date = frankfurt_set.loc[ind, "day"].split("-")
    year = int(date[0])
    month = int(date[1])
    day = int(date[2])
    hour = int(frankfurt_set.loc[ind, "time"].split(":")[0])
    frankfurt_set.loc[ind, "Zeitstempel"] = pd.Timestamp(year, month, day, hour)
```

```python
def get_date(time):
    # return str(time[0:4] + "-" + time[4:6] + "-" + time[6:8])
    return pd.Timestamp(int(time[0:4]), int(time[4:6]), int(time[6:8]), int(time[8:10]))
```

```python
print(get_date("201902010100"))
```

```python
data_TT["Zeitstempel"] = data_TT["Zeitstempel"].astype("str").apply(get_date)
```
